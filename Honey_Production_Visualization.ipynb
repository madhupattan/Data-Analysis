5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and
write your findings.

```python
sns.scatterplot(x=df['numcol'],y=df['prodvalue'],data=df)
plt.show()
```

Inference: 

    Majority of honey producing colonies have the production value between $0 to $2.Some colonies have production value as high     as $7.


6. Plot and understand the relationship between categorical variable 'year' and a numerical variable
'prodvalue' through boxplot, and write your findings

```python
plt.figure(figsize=(20,8))
sns.boxplot(df['year'],df['prodvalue'],data=df)
plt.show()
```

Inference:

For years 1998-2001, the prodvalue is the same.
From year 2002, the prodvalue is slightly increasing till 2012.
Year 2010 has the maximum prodvalue.

7. Visualize and understand the relationship between the multiple pairs of variables throughout different years
using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
var=df[['numcol','totalprod','prodvalue','year']]
sns.pairplot(data=var)
plt.show()
```

Inference:

'numcol','totalprod' and 'prodvalue' are highly correlated to each other.
There is a same relationship with 'year' and other variables.
